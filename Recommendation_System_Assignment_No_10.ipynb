Calculating Cosine Similarity between **Users**

```python
user1 = 1 - pairwise_distances( book_new.values, metric='cosine')
user1
```

```python
np.fill_diagonal(user1, 0)
```

```python
#Storing the results in a dataframe
user = pd.DataFrame(user1)
user
```

```python
#Setting the index and column names to user ids 
user.index = book.UserId.unique()
user.columns = book.UserId.unique()
```

**Considering the most similar customer**

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
def recommend(custID):
    simID = user.loc[custID].idxmax()
    simID_book = book[book['UserId'] == simID].Title
    custID_book = book[book['UserId'] == custID].Title
    return set(simID_book) - set(custID_book)
```

```python
recommend(507)
```

```python
recommend(276813)
```
